```python
# Como mencionado no arquivo infoDIM.md, uma imagem é composta por pixels
# Para processarmos uma imagem, precisamos primeiro carregá-la e acessar seus pixels
# Cada pixel em uma imagem é representado por um número que indica sua cor
# Para que possamos trabalhar com a imagem, precisamos lê-la no programa, e para isso, utilizamos a função cv.imread()
```

```python
img = cv.imread(img2_path) # Carrega a imagem
plt.imshow(img) # Exibe a imagem
print(img) # Exibe a matriz de pixels da imagem
```

```python
# Como podemos ver a cor branca é representada por [255, 255, 255] e a cor preta por [0, 0, 0]
```

```python
# Acessar um pixel específico
# Para acessar um pixel específico, basta passar as coordenadas do pixel na matriz da imagem
# A matriz da imagem é composta por 3 camadas, cada uma representando uma cor (RGB)
print(img[0, 0]) # Acessa o pixel na posição (0, 0)
print(img[0, 100]) # Acessa o pixel na posição (0, 100)
```

```python
img_copy = img.copy() # Cria uma cópia da imagem
img_copy[2,2] = [0, 0, 255] # Muda a cor do pixel na posição (2, 2) para azul
img_copy[2,3] = [0, 0, 255] # Muda a cor do pixel na posição (2, 3) para azul
img_copy[2:12, 2:12] = [0, 0, 255]  # Altera todos os pixels dessa área para azul
plt.imshow(img_copy) # Exibe a imagem
print(img.shape)
```

```python
img = cv.imread(img_path) # Lê a imagem
plt.imshow(img) # Mostra a imagem lida num plot
```

```python
# Quando lemos uma imagem com o OpenCV, ela é lida no formato BGR (Blue, Green, Red)
# Para converter para RGB, utilizamos o método cv.cvtColor()
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
# Para converter a imagem para escala de cinza, utilizamos o método cv.cvtColor() com a flag cv.COLOR_BGR2GRAY
img_gray = cv.cvtColor(img, cv.COLOR_BGR2GRAY)  # Converte a imagem para escala de cinza
plt.imshow(img_gray, cmap='gray')  # Mostra a imagem em escala de cinza
```

```python
# Aplicar uma máscara na imagem
# Uma máscara é uma matriz que contém valores 0 e 1
retval, img_mask = cv.threshold(img_gray, 200, 255, cv.THRESH_BINARY)  # Cria uma máscara binária
plt.imshow(img_mask, cmap='gray')  # Mostra a máscara
```

```python
#Inverter a máscara
img_mask_inv = cv.bitwise_not(img_gray)  
plt.imshow(img_mask_inv, cmap='gray')  
```

```python
# Essas técincas de manipulaçao de cor de imagens são muito utilizadas para reduzir a complexidade de processamento de imagens
# Entretanto, é importante lembrar que a perda de informação é grande, e que a imagem em escala de cinza não é adequada para todas as aplicações
# Para cada aplicação, é necessário avaliar qual a colocalidade de cor necessária, visando a melhor performance do algoritmo
```

```python
# Cortar imagem 
img_cropped = img[200:500, 200:800]  # Corta a imagem
img_cropped_rgb = cv.cvtColor(img_cropped, cv.COLOR_BGR2RGB) # Converte a imagem para RGB
plt.imshow(img_cropped_rgb)  # Mostra a imagem cortada
```

```python
# Transformações geométricas
# As transformações geométricas são utilizadas para modificar a forma de uma imagem
# As transformações mais comuns são a rotação, translação e escala
img = cv.imread("/Users/gabriel/MyProjectsGit/Computer Vision/Computer-Vision/2.ProcessamentoIMG/img.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)  

img_rotated_horz = cv.flip(img, 1)
img_rotated_vert = cv.flip(img, 0)
img_rotated_both = cv.flip(img, -1)

# Mostrar imagens 
plt.figure(figsize=(10, 5))
plt.subplot(141); plt.imshow(img); plt.title("Original")
plt.subplot(142); plt.imshow(img_rotated_horz); plt.title("Horizontal")
plt.subplot(143); plt.imshow(img_rotated_vert); plt.title("Vertical")
plt.subplot(144); plt.imshow(img_rotated_both); plt.title("Ambos")
```
